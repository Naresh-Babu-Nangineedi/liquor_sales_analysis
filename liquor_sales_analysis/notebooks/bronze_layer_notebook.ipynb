```python
# Load Bronze layer from uploaded CSV

import re
from pyspark.sql.functions import col

df_bronze = spark.read.format("csv") \
    .option("header","true") \
    .option("inferSchema","true") \
    .load("Files/Iowa_Liquor_Sales_20250827.csv")
```

```python
def sanitize_column_name(name):
    return re.sub(r'[^0-9a-zA-Z_]', '_', name)

df_bronze_clean = df_bronze
for original in df_bronze.columns:
    df_bronze_clean = df_bronze_clean.withColumnRenamed(original, sanitize_column_name(original))
```

```python
df_bronze_clean.write.mode("overwrite").saveAsTable("bronze_liquor_sales")
```
